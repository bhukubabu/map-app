```python
df0=df[df['cluster']==0]['TOTAL IPC CRIMES']
df1=df[df['cluster']==1]['TOTAL IPC CRIMES']
df2=df[df['cluster']==2]['TOTAL IPC CRIMES']
```

```python
c0_min,c0_max=df0.min(),df0.max()
c1_min,c1_max=df1.min(),df1.max()
c2_min,c2_max=df2.min(),df2.max()
```

```python
 print(c0_min,c0_max)
 print(c1_min,c1_max)
 print(c2_min,c2_max)
```

```python
import random
```

```python
def get_intensity(intensity,minn,maxx):
  return random.uniform(minn,maxx)
```

```python
for i,j in df.iterrows():
  if df['cluster'][i]==0:
    df['intse'][i]=get_intensity(df['TOTAL IPC CRIMES'][i],0,0.47)
  elif df['cluster'][i]==2:
    df['intse'][i]=get_intensity(df['TOTAL IPC CRIMES'][i],0.53,1.0)
  else:
    None
```

```python
df.head()
```

```python
df['intse']=df['intse'].apply(lambda x: get_intensity(x))
```

```python
def recluster(intensity):
  if intensity<=c1_max and intensity>=c1_min:
    return 1
  elif intensity<=c2_max and intensity>=c2_min:
    return 2
  elif intensity<=c0_max and intensity>=c0_min:
    return 0
  else:
    return 0
```

```python
df['cluster']=df['cluster'].apply(lambda x:recluster(x))
```

```python
df.to_csv('final_lat.csv')
```
